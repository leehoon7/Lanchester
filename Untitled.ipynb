```python
import math
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import matplotlib.pylab as pl
```

```python
def get_time_interval(alpha, R, beta, B):
    param_lambda = alpha * R + beta * B
    return np.random.exponential(1 / param_lambda)

def get_blue_win(alpha, R, beta, B):
    weight_b = beta * B
    weight_r = alpha * R
    blue_win = weight_b / (weight_b + weight_r)
    return np.random.binomial(1, blue_win)
```

```python
# initialization
np.random.seed(2022)
batch_num = 1000
run_time = 1000000
alpha, beta = 0.01, 0.01
R, B = 46, 40

# game info initialization
results_R = []  # dummy array
results_B = []
sim_time = []
done = []
game_len = []
```

```python
# simulation start
for _ in range(int(run_time / batch_num)):

    # game step info initialization (for batch)
    R_batch_all = [np.ones(batch_num) * R]
    B_batch_all = [np.ones(batch_num) * B]
    sim_time_batch_all = [np.zeros(batch_num)]
    done_batch_all = [np.array([False] * batch_num)]

    # game info initialization (for batch)
    R_batch = R_batch_all[0].copy()
    B_batch = B_batch_all[0].copy()
    sim_time_batch = sim_time_batch_all[0].copy()
    done_batch = done_batch_all[0].copy()

    # vectorized simulation
    while done_batch.sum() != batch_num:

        # 1 step for game
        next_time = get_time_interval(alpha, R_batch, beta, B_batch)
        blue_win = get_blue_win(alpha, R_batch, beta, B_batch)
        next_time[done_batch] = 0.0

        # game info update (only for valid game)
        sim_time_batch += next_time
        blue_win_game = np.logical_and(blue_win == 1, np.logical_not(done_batch))
        red_win_game = np.logical_and(blue_win == 0, np.logical_not(done_batch))
        R_batch[blue_win_game] -= 1
        B_batch[red_win_game] -= 1

        # masking for finished game
        done_batch[R_batch == 0] = True
        done_batch[B_batch == 0] = True

        # something
        R_batch_all.append(R_batch.copy())
        B_batch_all.append(B_batch.copy())
        sim_time_batch_all.append(sim_time_batch.copy())
        done_batch_all.append(done_batch.copy())

    results_R.append(R_batch_all)
    results_B.append(B_batch_all)
    sim_time.append(sim_time_batch_all)
    done.append(done_batch_all)
    game_len.append(len(done_batch_all))
```

```python
results_R = [np.stack(results_R_item, axis=-1) for results_R_item in results_R]
results_B = [np.stack(results_B_item, axis=-1) for results_B_item in results_B]
sim_time = [np.stack(sim_time_item, axis=-1) for sim_time_item in sim_time]
done = [np.stack(done_item, axis=-1) for done_item in done]
sim_time_max = max([sim_time_item.max() for sim_time_item in sim_time])
```

```python
def animate_(clock, ax, sim_time, results_R, results_B, batch_num, R, B):

    R_num = []
    B_num = []

    for batch_idx in range(int(run_time / batch_num)):

        cur_idx = (sim_time[batch_idx] <= clock).sum(-1) - 1
        R_num.append(results_R[batch_idx][np.arange(batch_num), cur_idx])
        B_num.append(results_B[batch_idx][np.arange(batch_num), cur_idx])

    R_num = np.concatenate(R_num)
    B_num = np.concatenate(B_num)

    count_num = np.zeros((R + 1, B + 1), dtype=np.int32)

    for r_i in range(R + 1):
        for b_i in range(B + 1):
            count_num[r_i, b_i] = np.logical_and(R_num == r_i, B_num == b_i).sum()
    
    count_num = np.clip(count_num / count_num.sum(), 0.0, 0.06)
    R_max, B_max = np.unravel_index(np.argmax(count_num, axis=None), count_num.shape)
    
    r_grid = np.linspace(0, R, R + 1)
    b_grid = np.linspace(0, B, B + 1)
    B_grid, R_grid = np.meshgrid(b_grid, r_grid)
    
    global wframes
    if wframes[0]:
        for wframe in wframes:
            try:
                ax.collections.remove(wframe)
            except:
                pass
    
    global max_all
    max_all.append([B_max, R_max, np.max(count_num, axis=None)])
        
    B_finished = count_num[0].copy()
    R_finished = count_num[:, 0].copy()
    
    if wframes[0]:
        wframes[3].set_data(np.zeros_like(R_finished), np.linspace(0, R, R+1))
        wframes[3].set_3d_properties(R_finished)
        wframes[4].set_data(np.linspace(0, B, B+1), np.zeros_like(B_finished))
        wframes[4].set_3d_properties(B_finished)
        wframes[5].set_data(np.array(max_all)[:, 0], np.array(max_all)[:, 1])
        wframes[5].set_3d_properties(np.array(max_all)[:, 2])
    else:
        wframes[3] = ax.plot(np.zeros_like(R_finished), np.linspace(0, R, R+1),
                             R_finished, linewidth=3, color='red')[0]
        wframes[4] = ax.plot(np.linspace(0, B, B+1), np.zeros_like(B_finished),
                             B_finished, linewidth=3, color='blue')[0]
        wframes[5] = ax.plot(np.array(max_all)[:, 0], np.array(max_all)[:, 1],
                             np.array(max_all)[:, 2], color='green', linewidth=2)[0]
        
    wframes[0] = ax.plot_wireframe(B_grid, R_grid, count_num,
                                   color='black', rstride=1, cstride=1, linewidth=0.5)
    
    fig.tight_layout()
    fig.suptitle('time : ' + str(clock))
    
    ax.xaxis.pane.set_edgecolor('grey')
    ax.yaxis.pane.set_edgecolor('grey')
    ax.xaxis.pane.fill = False
    ax.yaxis.pane.fill = False
    ax.zaxis.pane.fill = False
    
    ax.set_xlabel('Blue', fontsize=12)
    ax.set_ylabel('Red', fontsize=12)
    ax.set_zlabel('ratio', fontsize=12)
        
    ax.set_xticks(np.arange(0, B+1, 10))
    ax.set_yticks(np.arange(0, R+1, 10))
    ax.set_zticks(np.linspace(0.00, 0.06, 4))
    
    ax.view_init(elev=20., azim=240)
    
    return fig
```

```python
fig, ax = plt.subplots(ncols=1, figsize=(15, 9), subplot_kw={"projection":"3d"}, dpi=80)
wframes, max_all = [None for _ in range(6)], []
_ = animate_(100, ax, sim_time, results_R, results_B, batch_num, R, B)
```

```python
from functools import partial
from matplotlib import animation
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots(ncols=1, figsize=(15, 9), subplot_kw={"projection":"3d"}, dpi=80)

wframes, max_all = [None for _ in range(6)], []
num_frame = math.ceil(sim_time_max + 1)
animate = partial(animate_, ax=ax, sim_time=sim_time, 
                  results_R=results_R, results_B=results_B, 
                  batch_num=batch_num, R=R, B=B)
anim = animation.FuncAnimation(fig, animate, frames=300, interval=1)
anim.save('lanchester.gif', fps=10)
```
